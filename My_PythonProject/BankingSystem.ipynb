```python
import os

class Banking:
    Acc_info = 1
    
    def __init__(self):
        self.accounts = {}
        # Optionally, load existing accounts from file if needed

    def account_registration(self):
        file_path = f"C:/Users/Admin/Desktop/banking{Banking.Acc_info}.txt"
        with open(file_path, "w") as file:
            name = input("Enter the name: ")
            file.write(name + "\n")
            account_number = input("Enter the account number: ")
            file.write(account_number + "\n")
            password = input("Enter the password: ")
            file.write(password + "\n")
            account_type = input("Enter the type of account (Savings/Current): ")
            file.write(account_type + "\n")
            address = input("Enter the address: ")
            file.write(address + "\n")
            branch = input("Enter the home branch: ")
            file.write(branch + "\n")

        Banking.Acc_info += 1

        self.accounts[account_number] = {
            'Name': name,
            'Password': password,
            'Type of Account': account_type,
            'Address': address,
            'Branch': branch,
            'Balance': 0.0
        }
        print("Registration Completed")

    def Login(self):
        account_number = input("Enter the account number: ")
        password = input("Enter the password: ")

        if account_number in self.accounts:
            account = self.accounts[account_number]
            if password == account['Password']:
                print(f"Name: {account['Name']}")
                print(f"Account Number: {account_number}")
                print(f"Type of Account: {account['Type of Account']}")
                print(f"Address: {account['Address']}")
                print(f"Branch: {account['Branch']}")
                print(f"Balance: {account['Balance']:.2f}")
            else:
                print("Invalid password")
        else:
            print("Invalid account number")

    def check_balance(self):
        account_number = input("Enter the account number: ")
        if account_number in self.accounts:
            balance = self.accounts[account_number]['Balance']
            print(f"Your balance is {balance:.2f}")
        else:
            print("Invalid account number")

    def deposit(self):
        account_number = input("Enter the account number: ")
        if account_number in self.accounts:
            try:
                amount = float(input("Enter the amount to deposit: "))
                if amount <= 0:
                    print("Please enter a valid amount")
                else:
                    self.accounts[account_number]['Balance'] += amount
                    print(f"Deposit of {amount:.2f} successful")
            except ValueError:
                print("Invalid amount entered")
        else:
            print("Invalid account number")

    def withdrawal(self):
        account_number = input("Enter the account number: ")
        if account_number in self.accounts:
            try:
                amount = float(input("Enter the amount to withdraw: "))
                if amount <= 0:
                    print("Please enter a valid amount")
                elif amount > self.accounts[account_number]['Balance']:
                    print("Insufficient funds")
                else:
                    self.accounts[account_number]['Balance'] -= amount
                    print(f"Withdrawal of {amount:.2f} successful")
            except ValueError:
                print("Invalid amount entered")
        else:
            print("Invalid account number")

def main():
    user = Banking()

    while True:
        print("******************************")
        print("Welcome to Bank of Maharashtra")
        print("1: Registration")
        print("2: Login")
        print("3: Check Balance")
        print("4: Deposit Money")
        print("5: Withdraw Money")
        print("6: Exit")
        print("******************************")

        choice = input("Enter your choice (1/2/3/4/5/6): ")

        if choice == '1':
            user.account_registration()
        elif choice == '2':
            user.Login()
        elif choice == '3':
            user.check_balance()
        elif choice == '4':
            user.deposit()
        elif choice == '5':
            user.withdrawal()
        elif choice == '6':
            print("Thank you, visit again!")
            break
        else:
            print("Select a valid option from 1 to 6")

if __name__ == "__main__":
    main()
    file.close()
```
